```python
import asyncio
from playwright.async_api import async_playwright

async def scrape_thread_links(url: str):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        # Go to the Apple Discussions search page
        await page.goto(url)

        # Wait until results are fully loaded
        await page.wait_for_selector("a[href*='/thread/']", timeout=90000) # Increased timeout to 90 seconds

        # Extract all thread links
        thread_links = await page.eval_on_selector_all(
            "a[href*='/thread/']",
            "elements => elements.map(el => el.href)"
        )

        await browser.close()
        return thread_links
```

```python
with open("discussion.links", "w") as f :
    f.write("")
```

```python
async def get_links(last_page, query="iphone+trade+in", start_page=1) :
    links = []
    count = 0
    for page in range(start_page, last_page+1):
        print(f"Page {page}, Count {count}")
        url = f"https://discussions.apple.com/search?source=aml&origin=asc_serp&q={query}&page={page}&perPage=20&content=filterDiscussions&sortBy=relevance"
        links = await scrape_thread_links(url)

        prev = ""

        with open("discussion.links", "a+") as f :
            for link in links :
                if prev != link :
                    if "?" not in link :
                        f.write(link + "\n")
                        count += 1
                        prev = link
```

```python
await get_links(100, start_page=1)
```
